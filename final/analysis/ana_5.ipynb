Find the applications related to software and display on US map

```python
from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read CSV
df = pd.read_csv('../data/raw_data/h1b_petition.csv')
df.dropna()
```

```python
job_spot=df[df.JOB_TITLE.str.contains('SOFTWARE')==True]
#Get lat and lons from the jobs
lons=list(job_spot.lon.fillna(0))
lats=list(job_spot.lat.fillna(0))
```

```python
# create the figure and axes instances
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
m = Basemap(llcrnrlon=-119,llcrnrlat=22,urcrnrlon=-64,urcrnrlat=49,\
            resolution='c',area_thresh=1000.,projection='lcc',\
            lat_1=33,lat_2=45,lon_0=-95)
m.etopo()
m.drawcoastlines()
m.drawcountries(linewidth=2)
m.drawstates()

# draw parallels and meridians
parallels = np.arange(0.,80,20.)
m.drawparallels(parallels,labels=[1,0,0,1])
meridians = np.arange(10.,360.,30.)
m.drawmeridians(meridians,labels=[1,0,0,1])

# add points
x,y = m(lons, lats)
m.plot(x, y, 'o', markersize=2, color="pink")

ax.set_title('Related to Software job position (2011-2016)')
plt.show()
```
