Application prevailing wage distribution in a company and average wage in each state
    

```python
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np;
import string
%matplotlib inline
import seaborn as sns
```

```python
# Read CSV
df = pd.read_csv('../data/raw_data/h1b_petition.csv')
df.dropna()
df['CITY'],df['STATE'] = df['WORKSITE'].str.split(',', 2).str[0:2].str
df.head()
```

```python
df['STATE'].unique()
```

```python
wage=df[df['STATE']==' MASSACHUSETTS']
wage['EMPLOYER_NAME'].value_counts().head()
```

```python
# Find prevailing wage distribution of H1B petition in Wipro Limited
wage=df[df['EMPLOYER_NAME']=='WIPRO LIMITED']
wage['PREVAILING_WAGE'].value_counts().tolist
wage=wage[wage['PREVAILING_WAGE']<600000]
sns.distplot(wage['PREVAILING_WAGE']);
```

```python
# Find average prevailing wage in each state
dfR = df[['STATE','PREVAILING_WAGE']]
dfR.head()
```

```python
# Find top 20 prevaling wage states
dfR.groupby('STATE').mean().sort_values(by='PREVAILING_WAGE',ascending=True).head(20).plot(kind='barh')
```
