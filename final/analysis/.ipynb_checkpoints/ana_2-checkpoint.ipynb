Find the top 10 application states and jobs and top 10 jobs in each state

```python
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import pandas as pd
import string
%matplotlib inline
import seaborn as sns
```

```python
# Read CSV
df = pd.read_csv('../data/raw_data/h1b_petition.csv')
df.dropna()
df.head()
```

```python
df['CITY'],df['STATE'] = df['WORKSITE'].str.split(',', 2).str[0:2].str
```

```python
df.head()
```

```python
# Find top 10 most application states
df.groupby('STATE').size().sort_values(ascending=False).head(10).plot(kind='bar')
```

```python
# Find top 10 application Job
df.groupby(['SOC_NAME']).size().sort_values(ascending=False).head(10).plot(kind='bar')
```

```python
# Find top 10 application job in each states and write into csv
df['COUNTER'] =1 #initially, set that counter to 1.
dfUsing = df[['STATE','SOC_NAME','COUNTER']]
      
group_data = dfUsing.groupby(['STATE','SOC_NAME']).sum().reset_index() #sum function
g = group_data['COUNTER'].groupby(level=0, group_keys=False)
f = lambda x: x.sort('COUNTER', ascending=False).head(10)
sort2 = group_data.groupby('STATE', sort=False,as_index=False).apply(f)
topJob = sort2.reset_index()[['STATE','SOC_NAME','COUNTER']]
topJob.head()
topJob.to_csv('ana_2/Result.csv',index=False)
```
