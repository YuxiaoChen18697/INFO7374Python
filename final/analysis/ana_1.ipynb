Find top 10 petition employers in 5 years and the application trend of the top 5 employers

```python
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import pandas as pd
import string
%matplotlib inline
import seaborn as sns
```

```python
# Read CSV
df = pd.read_csv('../data/raw_data/h1b_petition.csv')
df.dropna()
df.head()
```

```python
# Top 10 employer applying H1B in 5 years
topEmployer = df['EMPLOYER_NAME'].value_counts().sort_values(ascending=False).head(10).plot(kind="bar")
```

```python
# H1B Application trend for Top Employers in last 5 years
topEm = ['INFOSYS LIMITED','TATA CONSULTANCY SERVICES LIMITED','WIPRO LIMITED',
                                 'DELOITTE CONSULTING LLP','IBM INDIA PRIVATE LIMITED']
df2=df[df['EMPLOYER_NAME'].isin(topEm)].groupby(['EMPLOYER_NAME','YEAR']).size().unstack()
df2.plot(kind='barh')
```
